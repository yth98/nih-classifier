```python
import os
import pandas as pd
import cv2
```

```python
def data_resize(train_img_folder = './data/data/images/', path = './data/ntu_final_2018/train.csv',
    resized_img_folder = './data/data/rez224/', reshaped_size = (224,224), max_image_cnt = 10002):
    if not os.path.exists(resized_img_folder):
        os.makedirs(resized_img_folder)
    
    # loading the dataset
    train_df = pd.read_csv(path, dtype="str")

    idx = train_df['Image Index']

    print("Training Images Resize")
    cnt = 0
    one_peercent_len = max_image_cnt // 100
    for img_id in idx:
        if  cnt % one_peercent_len == 0:
            print (cnt / one_peercent_len, "% done")
        if  cnt >= max_image_cnt:
            break

        img = cv2.imread(train_img_folder + img_id, cv2.IMREAD_GRAYSCALE)
        res = cv2.resize(img, dsize = reshaped_size)
        cv2.imwrite(resized_img_folder + img_id, res)
        cnt += 1
    print("Done Training Images Resizing")
```

```python
data_resize()
```
